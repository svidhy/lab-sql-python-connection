```python
import pandas as pd
import numpy as np

import pymysql                        # for getting data from a SQL database
from sqlalchemy import create_engine  # for establishing the connection and authentication

from getpass import getpass  # To get the password without showing the input
```

```python
password = getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
def rentals_month(engine, month, year):
    
    if (1<= month < 12):
        #check if month is a valid integer
        query = "SELECT *  FROM rental WHERE month(rental_date) = "+str(month)+" and YEAR(rental_date)= "+str(year)
        print(query)
    
        df = pd.read_sql_query(query,engine)
        return df
    else: #Any other value, the function ends.
        raise Exception('Please enter valid month')
    
```

```python
def rental_count_month(df, month, year):
    
    #Throw Exception if the Dataframe is 0
    if len(df)<1 :
        return df
        
    coln_name = "rentals_"
    
    if (month < 10):
        mm = str(month)
        mm = '0' + mm
    else:
        mm = str(month)
        
    coln_name = coln_name+mm+"_"+str(year) #creates the column name
        
    data = df.groupby(['customer_id'])['rental_id'].count().reset_index(name=coln_name)
    #data = data.to_frame()
    return data
```

```python
def compare_rentals(df1, df2):
    
    if (len(df1) < 1 or len(df2) < 1):
        raise Exception ('Invalid input')
    
    df_new = pd.merge(df1, df2, on='customer_id') #merge two dataframe on customer id
    df_new['difference'] = df_new.iloc[:,1] - df_new.iloc[:,2] #Difference between col1 and col2. Col0 is customer_id
    
    return df_new
```

```python
dataf1 = rentals_month(engine,5, 2005)
dataf1
```

```python
dataf2 = rentals_month(engine,8, 2005)
dataf2
```

```python
dataf1_count = rental_count_month(dataf1,5, 2005)
dataf1_count
```

```python
dataf2_count = rental_count_month(dataf2,8, 2005)
dataf2_count
```

```python
df_diff = compare_rentals(dataf1_count, dataf2_count)
df_diff
```
